# *italicized text* Categorical Data - Bar chart, Pie Chart

Step 1: Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load Dataset


```python
df = pd.read_csv("StudentsPerformance.csv")
```

Step 3: Categorical Summary (Univariate)

```python
print(df['gender'].value_counts())
print(df['gender'].value_counts(normalize=True) * 100)
```

Step 4: Bar Chart

```python
sns.countplot(x='gender', data=df)
plt.title("Bar Chart - Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Step 5: Pie Chart

```python
gender_counts = df['gender'].value_counts()
# Plot pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Gender Distribution")
plt.axis('equal') # Equal aspect ratio ensures pie is a circle
plt.show()
```

 Continuous Data â€“ Common Visualizations:


Step 1:import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

 Step 2: Load the Dataset

```python
df = pd.read_csv("StudentsPerformance.csv")
```

Step 3: Display First Few Rows

```python
print(df.head())
```

Step 4: Select a Continuous Variable

```python
math_scores = df['math score']
reading_scores = df['reading score']
writing_scores = df['writing score']
```

Step 5: Summary Statistics


```python
print("Math Score Summary:")
print(math_scores.describe())
print("\nReading Score Summary:")
print(reading_scores.describe())
print("\nWriting Score Summary:")
print(writing_scores.describe())
```

Step 6: Histogram (Distribution)


```python
plt.hist(df["math score"], bins=10, edgecolor='black')
plt.title("Histogram: Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

Step 7: Box Plot (Outliers and Spread)


```python
sns.boxplot(x=math_scores)
plt.title("Box Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

Step 8:Line Plot â€“ Math Score vs Index

```python
plt.plot(df.index, df["math score"], marker='o')
plt.title("Line Plot: Math Score by Student Index")
plt.xlabel("Student Index")
plt.ylabel("Math Score")
plt.show()
```

Step 9: Density Plot (Smooth Distribution)


```python
sns.kdeplot(math_scores, shade=True)
plt.title("Density Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

Step 10: Rug Plot (Exact Value Indicators)

```python
sns.rugplot(df["math score"], height=0.1)
plt.title("Rug Plot: Math Scores")
plt.xlabel("Math Score")
plt.show()
```

Sep 11:Strip Plot

```python
sns.stripplot(x="gender", y="math score", data=df, jitter=True)
plt.title("Strip Plot: Math Score vs Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

Step 12:Swarm Plot

```python
sns.swarmplot(x="test preparation course", y="writing score", data=df)
plt.title("Swarm Plot: Writing Score vs Test Prep")
plt.xlabel("Test Preparation Course")
plt.ylabel("Writing Score")
plt.show()
```
